# 找好兩個資料來源 找出其中交集 併成新資料
## 我想試圖找出車禍件數與降雨量是否有正相關

```python
import pandas as pd
df=pd.read_csv('107caraccident.csv')#檔名必須是英文,並用記事本改成utf-8 encode
df.head()
#先看一下原始資料長得怎樣
#這是北市107年所有的交通事故
```

```python
#由於資料量較大(50329筆)
#我們只要先只要一月的就好
df2=df[df['發生月']==1]
df2.head()
```

```python
df3=df2.sort_values(by='發生日').iloc[:,0:3]
df_accident_jan=df3.groupby('發生日').count().iloc[:,0:1]
df_accident_jan.columns={'當日車禍件數'}
df_accident_jan=df_accident_jan.reset_index(drop=True)
df_accident_jan
#我們想要計算一月每天的交通事故數量
```

```python
df_rain_year=pd.read_excel('rain.xlsx')
df_rain_year
#這是107年一整年每日降雨量
```

```python
#我們只需要一月的
df_rain_jan=df_rain_year.iloc[2:,1:2]
#做一點簡單的資料清洗
df_rain_jan.replace({'-':'0','T': '0','NaN':'0'},inplace=True)
df_rain_jan.columns={'一月降雨量'}
df_rain_jan=df_rain_jan.reset_index(drop=True)
df_rain_jan
```

```python
#今日做重要的事:合併這兩個資料
df_concat=pd.concat([df_rain_jan,df_accident_jan],axis=1)
df_concat
```
